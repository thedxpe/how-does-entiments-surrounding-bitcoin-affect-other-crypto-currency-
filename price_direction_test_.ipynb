```python
file_path = '/content/drive/MyDrive/Crypto_Features_with_Trends.csv'

df = pd.read_csv(file_path)
display(df.head())
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import chi2_contingency
from statsmodels.tsa.stattools import grangercausalitytests
import statsmodels.api as sm
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler

# -------------------
# 1. Load Data
# -------------------

df.sort_values("datetime", inplace=True)
df.reset_index(drop=True, inplace=True)

# -------------------
# 2. Create price returns & direction
# -------------------
coins = ["ETH", "SOL", "LTC", "AVAX", "ADA"]

for coin in coins:
    df[f"{coin}_Return"] = df[f"{coin}_Close"].pct_change()
    threshold = 0.001  # Define flat zone
    df[f"{coin}_Direction"] = np.where(df[f"{coin}_Return"] > threshold, 1,
                                np.where(df[f"{coin}_Return"] < -threshold, 0, 2))

# -------------------
# 3. Sentiment change & lags
# -------------------
df["Sentiment_Daily_Change"] = df["Sentiment_Daily"].diff()
df["Sentiment_5Day_Change"] = df["Sentiment_5Day_Cumulative"].diff()

for lag in [1, 2, 3]:
    df[f"Sentiment_Change_Lag{lag}"] = df["Sentiment_Daily_Change"].shift(lag)

# Drop NaN after feature creation
df.dropna(inplace=True)

# -------------------
# 4. Correlation & Rolling Correlation
# -------------------
for coin in coins:
    corr = df["Sentiment_Daily_Change"].corr(df[f"{coin}_Return"])
    print(f"{coin} | Correlation with BTC sentiment change: {corr:.4f}")

    # Rolling correlation
    rolling_corr = df["Sentiment_Daily_Change"].rolling(30).corr(df[f"{coin}_Return"])
    plt.figure(figsize=(10, 4))
    plt.plot(df["datetime"], rolling_corr, label=f"{coin} Rolling Corr (30d)")
    plt.axhline(0, color='black', linestyle='--', alpha=0.7)
    plt.title(f"Rolling Correlation between BTC Sentiment Change & {coin} Returns")
    plt.legend()
    plt.show()

# -------------------
# 5. Chi-Square Test (categorical association)
# -------------------
df["Sentiment_Dir"] = np.where(df["Sentiment_Daily_Change"] > 0, 1, 0)  # 1=Positive sentiment change, 0=Negative

for coin in coins:
    contingency = pd.crosstab(df["Sentiment_Dir"], df[f"{coin}_Direction"])
    chi2, p, _, _ = chi2_contingency(contingency)
    print(f"{coin} | Chi-square p-value: {p:.4f}")

# -------------------
# 6. Logistic Regression (predict price direction from sentiment change)
# -------------------
for coin in coins:
    X = df[["Sentiment_Daily_Change", "Sentiment_Change_Lag1", "Sentiment_Change_Lag2"]]
    y = df[f"{coin}_Direction"].replace(2, np.nan).dropna()  # Drop flats
    X = X.loc[y.index]

    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)

    log_reg = LogisticRegression()
    log_reg.fit(X_scaled, y)

    coef_df = pd.DataFrame({
        "Feature": X.columns,
        "Coefficient": log_reg.coef_[0]
    })
    print(f"\nLogistic Regression Coefficients for {coin}:\n", coef_df)

# -------------------
# 7. Granger Causality Tests (does sentiment lead returns?)
# -------------------
max_lag = 3
for coin in coins:
    print(f"\n=== Granger Causality: Sentiment -> {coin} Returns ===")
    grangercausalitytests(df[[f"{coin}_Return", "Sentiment_Daily_Change"]], maxlag=max_lag, verbose=True)

# -------------------
# 8. Regime-based Analysis
# -------------------
df["BTC_Return"] = df["BTC_Close"].pct_change()
df["BTC_Regime"] = pd.qcut(df["BTC_Return"], 3, labels=["Bear", "Sideways", "Bull"])

for regime in ["Bear", "Sideways", "Bull"]:
    print(f"\n=== Market Regime: {regime} ===")
    subset = df[df["BTC_Regime"] == regime]
    for coin in coins:
        corr = subset["Sentiment_Daily_Change"].corr(subset[f"{coin}_Return"])
        print(f"{coin} correlation in {regime}: {corr:.4f}")
```
