```python
file_path = '/content/drive/MyDrive/Crypto_Features_with_Trends.csv'

df = pd.read_csv(file_path)
display(df.head())
```

```python
df["SOL_Direction"] = (df["SOL_Close"].shift(-1) > df["SOL_Close"]).astype(int)
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from xgboost import XGBClassifier
from sklearn.metrics import classification_report, accuracy_score
import matplotlib.pyplot as plt
import shap

# -----------------------------
# 1. Parameters
# -----------------------------
lookback_hours = 48
forecast_horizon = 24

# -----------------------------
# 2. Load and Clean Data
# -----------------------------
# df = pd.read_csv("your_data.csv") # Uncomment and load your data if needed
df = df.copy()
df = df.dropna(subset=["SOL_Direction", "Sentiment_Daily", "BTC_Volatility_24h", "SOL_Volatility_24h"])

# -----------------------------
# 3. Create Lag Features
# -----------------------------
features_to_lag = [
    "Sentiment_Daily", "BTC_Volatility_24h", "BTC_Volume_Change",
    "SOL_Volatility_24h", "SOL_Volume_Change"
]

for col in features_to_lag:
    for lag in range(1, lookback_hours + 1):
        df[f"{col}_lag{lag}"] = df[col].shift(lag)

# -----------------------------
# 4. Create Interaction Features
# -----------------------------
df["BTC_MACD_x_Sentiment"] = df["BTC_MACD"] * df["Sentiment_Daily"]
df["BTC_Returns_x_Vol"] = df["BTC_Returns"] * df["BTC_Volatility_24h"]
df["SOL_Returns_x_Sentiment"] = df["SOL_Returns"] * df["Sentiment_Daily"]

# -----------------------------
# 5. Create Multi-class Target (Up=2, Down=0, Flat=1)
# -----------------------------
df["Target_SOL_Direction_10h"] = df["SOL_Returns_24h"].shift(-forecast_horizon)

def classify_direction(x, threshold=0.001):
    if x > threshold:
        return 2  # up
    elif x < -threshold:
        return 0  # down
    else:
        return 1  # flat

df["SOL_Direction_3Class"] = df["Target_SOL_Direction_10h"].apply(classify_direction)

# -----------------------------
# 6. Drop rows with NaNs
# -----------------------------
df_model = df.dropna()
df_model.replace([np.inf, -np.inf], np.nan, inplace=True)
df_model.dropna(inplace=True)

# -----------------------------
# 7. Define Feature Columns
# -----------------------------
selected_features = [
    # SOL features
    'SOL_Volume_Change', 'SOL_Volume_MA_3h', 'SOL_Volume_MA_24h', 'SOL_Volume_Ratio',
    'SOL_Volume_STD_24h', 'SOL_Volume_Trend_7d', 'SOL_Returns', 'SOL_Volatility_3h',
    'SOL_Volatility_24h', 'SOL_High_Low_Range', 'SOL_Range_STD_24h', 'SOL_Volatility_Trend_7d',
    'SOL_Returns_1h', 'SOL_Returns_24h', 'SOL_Cumulative_Return_3h', 'SOL_EMA_12', 'SOL_EMA_26',
    'SOL_MACD', 'SOL_Return_Trend_7d', 'Sentiment_Daily', 'Sentiment_5Day_Cumulative',
    'Sentiment_1Day_Lagged',

    # BTC features
    'BTC_Volume_Change', 'BTC_Volume_MA_3h', 'BTC_Volume_MA_24h', 'BTC_Volume_Ratio',
    'BTC_Volume_STD_24h', 'BTC_Volume_Trend_7d', 'BTC_Returns', 'BTC_Volatility_3h',
    'BTC_Volatility_24h', 'BTC_High_Low_Range', 'BTC_Range_STD_24h', 'BTC_Volatility_Trend_7d',
    'BTC_Returns_1h', 'BTC_Returns_24h', 'BTC_Cumulative_Return_3h', 'BTC_EMA_12', 'BTC_EMA_26',
    'BTC_MACD', 'BTC_Return_Trend_7d',

    # Interaction terms
    'BTC_MACD_x_Sentiment', 'BTC_Returns_x_Vol', 'SOL_Returns_x_Sentiment'
]

lag_features = [col for col in df_model.columns if "_lag" in col]
all_features = selected_features + lag_features

X = df_model[all_features]
y = df_model["SOL_Direction_3Class"].astype(int)

# -----------------------------
# 8. Train/Test Split (No Shuffling)
# -----------------------------
split_idx = int(len(X) * 0.8)
X_train, X_test = X.iloc[:split_idx], X.iloc[split_idx:]
y_train, y_test = y.iloc[:split_idx], y.iloc[split_idx:]

# -----------------------------
# 9. Scale Features
# -----------------------------
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# -----------------------------
# 10. Train XGBoost Model
# -----------------------------
model = XGBClassifier(
    objective='multi:softmax',
    num_class=3,
    n_estimators=200,
    max_depth=6,
    learning_rate=0.05,
    subsample=0.8,
    colsample_bytree=0.8,
    random_state=42
)

model.fit(X_train_scaled, y_train)

# -----------------------------
# 11. Evaluate Model
# -----------------------------
y_pred = model.predict(X_test_scaled)

print("\nüîç Classification Report:")
print(classification_report(y_test, y_pred))
print("\n‚úÖ Accuracy:", accuracy_score(y_test, y_pred))

# -----------------------------
# 12. Feature Importance Plot
# -----------------------------
importances = model.feature_importances_
indices = np.argsort(importances)[::-1]

plt.figure(figsize=(14, 6))
plt.title("Top 25 Feature Importances")
plt.bar(range(25), importances[indices[:25]])
plt.xticks(range(25), X.columns[indices[:25]], rotation=90)
plt.tight_layout()
plt.show()

# -----------------------------
# 13. SHAP Analysis (Optional)
# -----------------------------
explainer = shap.Explainer(model)
shap_values = explainer(X_test_scaled)
shap.summary_plot(shap_values, X_test, max_display=25)
```
